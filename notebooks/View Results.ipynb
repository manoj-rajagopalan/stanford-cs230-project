```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
%matplotlib inline

import os
import os.path
```

```python
image_index = 161

# test image
project_dir = os.path.join(os.environ['HOME'], 'proj', 'stanford-cs230-project')
results_dir = os.path.join(project_dir, 'results')
exp_results_dir = os.path.join(results_dir, 'test1')
result_img_dir = os.path.join(exp_results_dir, 'images')
image_name = 'disparity_{}.png'.format(image_index)
image = cv.imread(os.path.join(result_img_dir, image_name))
assert image is not None

# ground-truth disparity
dataset_dir = 'kitti_2015'
gt_image1_path = os.path.join(project_dir, dataset_dir, 'training', 'disp_noc_0', '{:06d}_10.png'.format(image_index))
gt_image2_path = os.path.join(project_dir, dataset_dir, 'training', 'disp_noc_1', '{:06d}_10.png'.format(image_index))
gt_image1 = cv.imread(gt_image1_path)
gt_image2 = cv.imread(gt_image2_path)

# U-net filtered image
filtered_img_path = '/cs230-datasets/disparityFilter/results/gord_test2/images/{:06d}_10.png'.format(image_index)
filtered_img = cv.imread(filtered_img_path)

# PSM image
psm_img_path = '/cs230-datasets/PSMNet/00/submission/{:06d}_10.png'.format(image_index)
psm_img = cv.imread(psm_img_path)

dpi = plt.rcParams['figure.dpi']
figsize = (3 * gt_image1.shape[0] / dpi, 3 * gt_image1.shape[1] / dpi)
plt.figure(figsize = figsize)
plt.subplot(1,2,1)
plt.imshow(gt_image1)
plt.subplot(1,2,2)
plt.imshow(gt_image2)
plt.figure(figsize = figsize)
plt.imshow(image)
plt.figure(figsize = figsize)
plt.imshow(filtered_img)
plt.figure(figsize = figsize)
plt.imshow(psm_img)
```
